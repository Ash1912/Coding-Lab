```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.rand(20)
y=np.random.rand(20)
plt.scatter(x,y)
plt.show()
```

```python
x=np.arange(0, 2, 0.1)
y1=np.log(1+2*x)
y2=x**2
plt.plot(x, y1, label='log')
plt.plot(x, y2, 
         c='g', ls='--', lw=2,
         marker='s', mfc='r', label='power')
plt.scatter(x, y, label='scatter1')
plt.scatter(x, y+1, label='scatter2', marker='d')
plt.scatter(x, y+2, label='scatter3',marker='^')
plt.title('Some Functions')
plt.xlabel('x')
plt.ylabel('f(x)')
# plt.legend()
plt.legend(loc='upper left')
plt.show()
```
